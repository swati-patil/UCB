```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#total number of schools
student_passed_maths = school_data_complete.loc[(school_data_complete["math_score"] >= 70), :]
percentage_stu = (len(student_passed_maths) * 100) / len(school_data_complete)
student_passed_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70), :]
percentage_stu_Reading = (len(student_passed_reading) * 100) / len(school_data_complete)
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
total_budget = school_data_complete["budget"].sum()
dist_summary = {}
dist_summary = [{"Total Schools" : len(school_data_complete["school_name"].unique()), 
                "Total Students" : len(school_data_complete["student_name"]),
                "Total Budget" : total_budget,
                "Average Math Score" : avg_math_score,
                "Average Reading Score" : avg_reading_score,
                "% Passing Math" : percentage_stu,
                "% Passing Reading" : percentage_stu_Reading,
                "% Overall Passing Rate" : (avg_math_score + avg_reading_score)/2}]
dist_summary_df = pd.DataFrame(dist_summary)
dist_summary_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
maths_data_filtered = school_data_complete[school_data_complete['math_score'] >=70]
reading_data_filtered = school_data_complete[school_data_complete['reading_score'] >=70]
grouped_school_data = school_data_complete.groupby(['school_name'])
#grouped_school_data['school_name'].count().head()
grouped_school_data['math_score', 'reading_score', 'budget'].mean()
#print (grouped_school_data['budget'].mean()/grouped_school_data['school_name'].count())
summary_by_school = pd.DataFrame(grouped_school_data["school_name"].count())
summary_by_school['School Type'] = grouped_school_data['type'].unique()
summary_by_school['Total School Budget'] = grouped_school_data['budget'].mean()
summary_by_school['Per Student Budget'] = grouped_school_data['budget'].mean()/grouped_school_data['school_name'].count()
summary_by_school['Average Math Score'] = grouped_school_data['math_score'].mean()
summary_by_school['Average Reading Score'] = grouped_school_data['reading_score'].mean()
maths_data_filtered = maths_data_filtered.groupby(['school_name'])
summary_by_school['% Maths Passed'] = (maths_data_filtered['math_score'].count() * 100)/grouped_school_data['school_name'].count()
reading_data_filtered = reading_data_filtered.groupby(['school_name'])
summary_by_school['% Reading Passed'] = (reading_data_filtered['math_score'].count() * 100)/grouped_school_data['school_name'].count()
summary_by_school['Overall Passing Rate'] = (summary_by_school['% Maths Passed'] + summary_by_school['% Reading Passed'])/2
summary_by_school.head()
```

```python
top_perf_schools = summary_by_school.sort_values(by=['Overall Passing Rate'], ascending=False)
top_perf_schools = top_perf_schools.round(2)
top_perf_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_perf_schools = summary_by_school.sort_values(by=['Overall Passing Rate'], ascending=True)
bottom_perf_schools = bottom_perf_schools.round(2)
bottom_perf_schools.head()
```
